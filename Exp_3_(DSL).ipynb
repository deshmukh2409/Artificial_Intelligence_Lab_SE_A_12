```python
import numpy as np

# Define the coefficient matrix A
A = np.array([[2, 3], [4, 5]])

# Define the constant vector B
B = np.array([10, 20])

# Solve the system of equations Ax = B for x
solution = np.linalg.solve(A, B)

print("Solution to the system of equations:")
print(f"x = {solution[0]:.2f}")
print(f"y = {solution[1]:.2f}")
```
