```python
import numpy as np

# Define the coefficient matrix A
A = np.array([
    [2, 3],
    [4, 5]
])

# Define the constants vector B
B = np.array([
    [10],
    [20]
])

print("Coefficient Matrix A:")
print(A)
print("\nConstants Vector B:")
print(B)

# Solve the system of linear equations Ax = B
solution = np.linalg.solve(A, B)

print("\nSolution (x, y):")
print(solution)

# You can access x and y individually
x = solution[0][0]
y = solution[1][0]

print(f"\nx = {x:.2f}")
print(f"y = {y:.2f}")
```

```python
import numpy as np

# Create a new non-singular 3x3 square matrix
new_matrix = np.array([
    [1, 2, 0],
    [0, 1, 0],
    [0, 0, 1]
])

print("New 3x3 Matrix:")
print(new_matrix)

# Compute the determinant of the new matrix
new_determinant = np.linalg.det(new_matrix)
print(f"\nDeterminant of the new matrix: {new_determinant}")

# Compute the inverse of the new matrix
if new_determinant != 0:
    inverse_matrix = np.linalg.inv(new_matrix)
    print("\nInverse of the new matrix:")
    print(inverse_matrix)
else:
    print("\nThis matrix is also singular and does not have an inverse.")
```
