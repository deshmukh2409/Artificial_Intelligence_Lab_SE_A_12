### Palindrome Number Checker

This code defines a function `is_palindrome_number(number)` that checks if a given integer is a palindrome. A number is a palindrome if it reads the same forwards and backward.

````python
def is_palindrome_number(number):
    """Checks if a given number is a palindrome."""
    if not isinstance(number, int):
        return "Input must be an integer."
    if number < 0:
        return "Palindrome check is typically for non-negative numbers."

    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the string with its reverse
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome number."
    else:
        return f"{number} is not a palindrome number."

# Example usage:
num1 = 121
print(is_palindrome_number(num1))

num2 = 12321
print(is_palindrome_number(num2))

num3 = 12345
print(is_palindrome_number(num3))

num4 = 0
print(is_palindrome_number(num4))

# Example with invalid input:
num5 = -121
print(is_palindrome_number(num5))

num6 = 12.34
print(is_palindrome_number(num6))

num7 = "abc"
print(is_palindrome_number(num7))
````

### Odd or Even Number Identifier

This code defines a function `check_odd_even(number)` that determines whether a given integer is odd or even.

````python
def check_odd_even(number):
    """Checks if a given number is odd or even."""
    if not isinstance(number, int):
        return "Input must be an integer."
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 4
print(check_odd_even(num1))

num2 = 7
print(check_odd_even(num2))

num3 = 0
print(check_odd_even(num3))

# Example with invalid input:
num4 = 3.14
print(check_odd_even(num4))

num5 = "hello"
print(check_odd_even(num5))
````

### Multiplication Table Generator

This code defines a function `display_multiplication_table(number, limit)` that prints the multiplication table for a given `number` up to a specified `limit`.

````python
def display_multiplication_table(number, limit=10):
    """Displays the multiplication table for a given number up to a specified limit."""
    if not isinstance(number, (int, float)):
        return "Input must be a number."
    if not isinstance(limit, int) or limit < 1:
        return "Limit must be a positive integer."

    print(f"Multiplication Table for {number} up to {limit}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage:
display_multiplication_table(7)

print("\n") # Add a newline for better readability between examples
display_multiplication_table(12, limit=5)

print("\n")
display_multiplication_table(3.5, limit=3)

# Example with invalid input:
print("\n")
display_multiplication_table("hello")

print("\n")
display_multiplication_table(5, limit=0)
````

### Factorial Calculator

This code defines a function `factorial(n)` that calculates the factorial of a non-negative integer `n`. The factorial of a non-negative integer `n` is the product of all positive integers less than or equal to `n`.

````python
def factorial(n):
    """Calculates the factorial of a non-negative integer."""
    if not isinstance(n, int) or n < 0:
        return "Input must be a non-negative integer."
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage:
number = 5
print(f"The factorial of {number} is: {factorial(number)}")

number = 0
print(f"The factorial of {number} is: {factorial(number)}")

number = 7
print(f"The factorial of {number} is: {factorial(number)}")

# Example with invalid input:
number = -3
print(f"The factorial of {number} is: {factorial(number)}")

number = 3.5
print(f"The factorial of {number} is: {factorial(number)}")
````

## Understand Expert Systems

### What is an Expert System?

An **Expert System** is a computer program that emulates the decision-making ability of a human expert. It is designed to solve complex problems and provide decision support in a specific domain by reasoning through bodies of knowledge, represented primarily as `if-then` rules, rather than through conventional procedural code.

### Key Components of an Expert System:

1.  **Knowledge Base:**
    *   **Role:** This component stores all the domain-specific knowledge, facts, and rules that the expert system uses for problem-solving. It's essentially the system's long-term memory, containing both factual knowledge (e.g., definitions, properties) and heuristic knowledge (e.g., rules of thumb, strategies).

2.  **Inference Engine:**
    *   **Role:** This is the 'brain' of the expert system. It processes the knowledge stored in the knowledge base and uses logical reasoning to draw conclusions, make deductions, or identify solutions. It typically employs techniques like forward chaining (data-driven) or backward chaining (goal-driven) to navigate the rules and facts.

3.  **User Interface:**
    *   **Role:** This component provides a means for the user to interact with the expert system. It allows users to input problems or queries, receive questions or requests for additional information from the system, and view the system's conclusions, recommendations, and explanations for its reasoning.

### How Expert Systems Function for Decision-Making:

Expert systems facilitate decision-making through a cyclical process:

1.  **User Interaction:** A user presents a problem or query to the system via the **User Interface**.
2.  **Inference:** The **Inference Engine** takes this input and begins to process it, searching through the **Knowledge Base** for relevant facts and rules. It applies its reasoning mechanisms (e.g., forward or backward chaining) to match user input with rules and deduce new facts or potential solutions.
3.  **Knowledge Retrieval:** As the inference engine reasons, it continuously consults the **Knowledge Base** to retrieve necessary information, apply rules, and update its understanding of the problem.
4.  **Problem Solving & Explanation:** Once a conclusion is reached or a recommendation is formulated, the **Inference Engine** passes this information back to the **User Interface**. The user interface then displays the solution and often provides an explanation of the reasoning process, justifying how the system arrived at its decision. This explanation capability is crucial for user trust and understanding.

## Design a Rule-Based System

### Subtask:
Outline a basic architecture for a rule-based expert system, defining how 'facts' and 'rules' will be structured, and the concept of an 'inference engine'.


### Subtask: Outline a basic architecture for a rule-based expert system

#### 1. Representation of 'Facts'

Facts will be represented as a simple list of strings. Each string will describe a single piece of information or observation. This approach keeps the fact base straightforward and easily extensible. For more complex scenarios, a dictionary or object-based representation could be used, but for a basic system, strings are sufficient.

**Example Fact Base:**

```python
facts = [
    "animal_is_mammal",
    "animal_eats_meat",
    "animal_has_stripes"
]
```

#### 2. Structure of 'Rules'

Each rule will be represented as a dictionary with two key-value pairs: `condition` and `consequent`. The `condition` will be a list of strings, where each string represents a fact that must be present in the fact base for the rule to fire. The `consequent` will be a single string representing a new fact to be added to the fact base (or an action to be taken) if the condition is met.

**Example Rule Structure:**

```python
rules = [
    {
        "condition": ["animal_is_mammal", "animal_eats_meat", "animal_has_stripes"],
        "consequent": "animal_is_tiger"
    },
    {
        "condition": ["animal_is_bird", "animal_can_fly"],
        "consequent": "animal_is_eagle"
    }
]
```

#### 3. Concept of the 'Inference Engine'

The inference engine is the core component that processes the facts and rules to derive new conclusions. Its high-level function is to iteratively apply rules whose conditions are satisfied by the current set of facts, adding new facts to the fact base until no more rules can be fired or a specific goal is reached.

For this basic system, the inference engine will employ a **forward chaining** reasoning approach. This means it will start with the known facts and work forwards, applying rules to deduce new facts. The process will be as follows:

1.  **Initialize:** Start with an initial set of known facts.
2.  **Match:** Iterate through all defined rules and check if the `condition` of any rule is fully satisfied by the current `facts` (i.e., all facts in the rule's condition list are present in the current fact base).
3.  **Fire:** For each rule whose condition is met, add its `consequent` to the fact base. Ensure that duplicate facts are not added.
4.  **Loop:** Repeat steps 2 and 3 until no new facts can be added in an iteration, indicating that all possible conclusions have been drawn from the initial facts and rules.

**High-level process for the Inference Engine:**

```
Input: Initial Facts, Rules
Output: Derived Facts

1. While new facts are being added:
2.   For each Rule in Rules:
3.     If all facts in Rule.condition are in CurrentFacts:
4.       Add Rule.consequent to CurrentFacts (if not already present).
5. Return CurrentFacts
```

## Implement the Core System

### Subtask:
Develop Python code for the inference engine. This engine will process a set of given facts against a knowledge base of rules to deduce conclusions or recommendations.


**Reasoning**:
The subtask requires implementing an inference engine in Python, so I will provide a code block that defines the `inference_engine` function as described in the instructions.



````python
def inference_engine(initial_facts, rules):
    """
    Processes a set of given facts against a knowledge base of rules to deduce conclusions.

    Args:
        initial_facts (list): A list of initial facts (strings).
        rules (list): A list of dictionaries, where each dictionary represents a rule.
                      Each rule dictionary must have 'condition' (list of strings)
                      and 'consequent' (string) keys.

    Returns:
        set: The set of all facts, including initial and deduced facts.
    """
    current_facts = set(initial_facts)
    facts_changed = True

    while facts_changed:
        facts_changed = False
        for rule in rules:
            # Check if all conditions for the rule are present in current_facts
            if set(rule['condition']).issubset(current_facts):
                consequent = rule['consequent']
                # If the consequent is not already in current_facts, add it
                if consequent not in current_facts:
                    current_facts.add(consequent)
                    facts_changed = True

    return current_facts

# Example Usage:
# Define some initial facts
initial_facts_example = ['has_fur', 'is_mammal']

# Define some rules
rules_example = [
    {'condition': ['has_fur'], 'consequent': 'is_animal'},
    {'condition': ['is_mammal'], 'consequent': 'breathes_air'},
    {'condition': ['is_animal', 'breathes_air'], 'consequent': 'is_living_thing'},
    {'condition': ['is_mammal', 'breathes_air'], 'consequent': 'is_warm_blooded'},
    {'condition': ['is_warm_blooded', 'is_animal'], 'consequent': 'vertebrate'}
]

# Run the inference engine
deduced_facts = inference_engine(initial_facts_example, rules_example)
print(f"Initial Facts: {initial_facts_example}")
print(f"Rules Used: {rules_example}")
print(f"Deduced Facts: {sorted(list(deduced_facts))}")

print("\n--- Another Example ---")
initial_facts_weather = ['is_raining', 'is_cold']
rules_weather = [
    {'condition': ['is_raining'], 'consequent': 'take_umbrella'},
    {'condition': ['is_cold'], 'consequent': 'wear_coat'},
    {'condition': ['take_umbrella', 'wear_coat'], 'consequent': 'stay_dry_and_warm'}
]

deduced_weather_facts = inference_engine(initial_facts_weather, rules_weather)
print(f"Initial Facts: {initial_facts_weather}")
print(f"Rules Used: {rules_weather}")
print(f"Deduced Weather Facts: {sorted(list(deduced_weather_facts))}")
````
